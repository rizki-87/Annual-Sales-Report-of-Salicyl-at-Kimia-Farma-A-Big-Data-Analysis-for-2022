# Data Transformation

```python
# replace with the path to the Excel file
file_path = ('/content/drive/MyDrive/PORTOFOLIO/RakaminxKimiaFarma/kimiafarmaxrakamin.xlsx')  # Sesuaikan dengan lokasi file Anda di Google Colab

# teading Excel files
excel_data = pd.ExcelFile(file_path)

# the names of the sheets found in your file
sheet_names = ['penjualan', 'pelanggan', 'barang',
               'penjualan_ds', 'pelanggan_ds', 'barang_ds', 'sandbox']

# loop through each sheet and save as a CSV file.
for sheet_name in sheet_names:
    df = pd.read_excel(excel_data, sheet_name=sheet_name)
    # Gantikan '/content/' dengan path yang diinginkan
    csv_file_path = f'/content/{sheet_name}.csv'
    df.to_csv(csv_file_path, index=False)
    print(f'Saved {csv_file_path}')
```
